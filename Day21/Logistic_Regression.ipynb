# Logistics Analysis of Bank Loan

```python
import numpy as np
import pandas as pd
```

```python
dt = pd.read_excel('Bank_Personal_Loan_Modelling.xlsx',sheet_name=1)
```

```python
dt.head()
```

```python
dt.shape
```

```python
dt.info()
```

```python
dt.describe().T
```

```python
import statsmodels.api as sm
```

```python
y = pd.DataFrame(dt['Personal Loan'],columns=['Personal Loan'])
```

```python
x = dt.drop(['Personal Loan'],axis=1)
```

```python
x1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y, x1)
```

```python
result = Logistic.fit()
```

```python
result.summary()
```

## Inference
### Most significant features
* Income
* Family
* Education
* CD Account
* Online
* CreditCard
### Significant features
* CCAvg
* Securities Account

# Logistic Analysis on Atrittion

```python
dt1 = pd.read_csv('general_data.csv')
```

```python
dt1.head()
```

```python
dt1.shape
```

```python
dt1.info()
```

```python
dt1['Attrition'] = dt1['Attrition'].map({'Yes' :1,'No' : 0})
```

```python
dt1.head()
```

```python
dt1.describe().T
```

```python
dt1 = dt1.drop(['EmployeeCount','EmployeeID','Over18','StandardHours'],axis=1)
```

```python
dt1.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
le = LabelEncoder()
```

```python
dt1 = dt1.dropna()
```

```python
dt1.BusinessTravel = le.fit_transform(dt1.BusinessTravel)
dt1.Department = le.fit_transform(dt1.Department)
dt1.EducationField = le.fit_transform(dt1.EducationField)
dt1.Gender = le.fit_transform(dt1.Gender)
dt1.JobRole = le.fit_transform(dt1.JobRole)
dt1.MaritalStatus = le.fit_transform(dt1.MaritalStatus)
```

```python
dt1.head()
```

```python
dt1.isnull().sum()
```

```python
x = dt1.drop(['Attrition'], axis=1)
y = dt1.Attrition
```

```python
X1 = sm.add_constant(x)
```

```python
Log_reg = sm.Logit(y, X1)
```

```python
result = Log_reg.fit()
```

```python
result.summary()
```

## Inference
### Following are most significant features
* Age
* Marital Status
* NumCompaniesWorked
* TotalWorkingYears
* TrainingTimesLastYear
* YearsSinceLastPromotion
* YearsWithCurrManager
### Following are significant features
* Department
* EducationField
* JobRole
